### XML

Extensible Markup Language (XML) es un lenguaje que extiende las funcionalidades de markup y nos permite tener un documento con reglas definidas para que tanto humanos como máquinas puedan entenderlo. 

Así como HTML tiene secciones definiadas por etiquetas de inicio y final. Estas secciones se llaman elementos. Elemento de más alto nivel se llama root, contiene a todos los demás. 

### RSS

Realy Simple Syndication

 Se utiliza para difundir información actualizada frecuentemente a usuarios que se han suscrito a la fuente de contenidos. El formato permite distribuir contenidos sin necesidad de un navegador, utilizando programas llamados agregadores de noticias, diseñados para leer contenidos RSS, tales como Mozilla Firefox, Thunderbird o Akregator, entre otros. A pesar de eso, es posible utilizar el mismo navegador para ver los contenidos RSS.

```python
import requests
import xml.etree.ElementTree as ET

import pandas as pd 

from bs4 import BeautifulSoup
```

```python
resp = requests.get('https://www.inegi.org.mx/rss/noticias/xmlfeeds?p=2,1')
print(resp.status_code)
```

```python
noticias_inegi = resp.content
print(noticias_inegi)
```

```python
with open('noticias_inegi_20210517.xml', 'wb') as fw:
    fw.write(noticias_inegi)
```

```python
tree = ET.parse('noticias_inegi_20210517.xml')
tree
```

```python
tree = ET.fromstring(noticias_inegi)
tree
```

```python
tree.tag
```

```python
tree.attrib
```

```python
for elem in tree:
    for subel in elem:
        for ssubel in subel:
            print(ssubel.text)
```

```python
tree = ET.parse('noticias_inegi_20210517.xml')
tree
```

```python
root = tree.getroot()
```

```python
root[0]
```

```python
root[1]
```

```python
for child in root:
    for grandchild in child:
        print(grandchild.tag, grandchild.attrib)
```

```python
for child in root:
    for grandchild in child:
        for grangrandchild in grandchild:
            print(grangrandchild.tag, grangrandchild.attrib)
```

```python
elements = [el.tag for el in root.iter()]
elements
```

```python
for title in root.iter('title'):
    print(title.text)
```

```python
content = requests.get("https://www.reforma.com/libre/estatico/rss/").content
content
```

```python
soup = BeautifulSoup(content, 'xml')
```

```python
soup
```

```python
rss_feeds_categories = soup.select('div.btnRSS a')

rss_feeds_categories
```

```python
rss_feeds_categories = [feed['href'] for feed in rss_feeds_categories]
```

```python
rss_feeds_categories
```

```python
import re

rss_feed_categories_clean = [re.findall(r'(?<=rss).*$', cat)[0] for cat in rss_feeds_categories]
```

```python
rss_feed_categories_clean
```

```python
url = 'https://www.reforma.com/rss'

rss_categories = [url+rss for rss in rss_feed_categories_clean]
```

```python
rss_categories
```

```python
content = requests.get(rss_categories[0]).content
soup = BeautifulSoup(content, 'xml')
soup
```

```python
content = requests.get(rss_categories[0]).content
root = ET.fromstring(content)
root
```

```python
titles = soup.find_all('title')
titles = [title.text for title in titles]
titles
```

```python
titles = [title.text for title in root.iter('title')]
titles
```

```python
titles = titles[2:]
titles
```
