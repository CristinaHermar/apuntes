```python
#UNION ALL SQL
df_final = df_01.append(data_frames)
```

```python
nulos = df_final.isnull().sum()
```

```python
df_no_nulos = df_final.drop(nulos[nulos>40000].index, axis = 1)
```

```python
df_no_nulos.head(3)
```

```python
df_no_nulos['nombre_act_limpio'] = df_no_nulos['nombre_act'].str.lower()
```

```python

# puntuacion = [".", "," , "-", "\n", ":", ";",'"', "'", "  ",
#                 "!", "¡", "¿", "?", "(", ")", "<", "«", "»", "~"]
# for p in puntuacion:
#   df_no_nulos['nombre_act_limpio'] = df_no_nulos['nombre_act'].str.replace(p,"")
```

```python
puntuacion = [".", "," , "-", "\n", ":", ";",'"', "'", "  ",
                "!", "¡", "¿", "?", "(", ")", "<", "«", "»", "~"]
for p in puntuacion:
  df_no_nulos['nombre_act_limpio'] = df_no_nulos['nombre_act_limpio'].str.replace(p,"")
```

```python
df_no_nulos['nombre_act'].unique().tolist()
```

```python
df_no_nulos['nombre_act'][df_no_nulos['nombre_act'].str.contains("humanidades")].unique().tolist()
```

```python
df_no_nulos['nombre_act_limpio'][df_no_nulos['nombre_act_limpio'].str.contains("humanidades")].unique().tolist()
```

```python
df_no_nulos['nombre_act_limpio'] = df_no_nulos['nombre_act_limpio'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
df_no_nulos['nombre_act_limpio'][df_no_nulos['nombre_act_limpio'].str.contains("humanidades")].unique().tolist()
```

```python
df_no_nulos['nombre_act_limpio'].value_counts()
```

```python
df_conteo = df_no_nulos.groupby(['nombre_act_limpio']).agg({'id':'count'}).reset_index()
```

```python
df_conteo.head(10)
```

```python
rows_to_remove = list(df_conteo[df_conteo['id']<100].index)
df_conteo_limpio = df_conteo.drop(rows_to_remove, axis=0)
df_conteo_limpio.head(10)
```

```python
df_top_10 = df_conteo_limpio.nlargest(10,'id')
df_top_10['class'] = 'top_10'
df_top_10.head(5)
```

```python
df_low_10 = df_conteo_limpio.nsmallest(10,'id')
df_low_10['class'] = 'low_10'
df_low_10.head(5)
```

```python
df_tops = df_low_10.append(df_top_10)
df_tops.head(2)
```

```python
df_no_nulos.head(2)
```

```python
df_final = pd.merge(df_no_nulos, df_tops, how = 'left', on =['nombre_act_limpio'])
df_final[df_final['class'].notnull()]
```

# Pivot Tables

```python
df_liquors = pd.read_csv('/content/drive/MyDrive/Databases/Week 9 - Wednesday/database_liquors.csv')
```

```python
df_liquors.head(2)
```

```python
df_liquors.pivot_table(index=['category_name','item_description'], aggfunc = np.sum)
```

```python
df_liquors.pivot_table(index=['category_name','item_description'], values= ['sale_usd','bottles_sold'],aggfunc = np.sum)
```

```python
df_liquors.pivot_table(index=['category_name','item_description'], values= ['sale_usd','bottles_sold'], columns=['county'],aggfunc = np.sum, fill_value=0)
```

```python
df_liquors.groupby(['category_name','item_description','county']).agg({'sale_usd':'sum','bottles_sold':'sum'})
```
