#Matplotllib

```python
df_liquors['semana'] = df_liquors['date'].dt.week #Series.dt.week >> The week ordinal of the year.
#en la página y aquí dicen que usemos dt.isocalendar().week
```

```python
df_liquors.head(2)
```

```python
df_liquors['date'].dt.year.unique()
```

```python
df_week = df_liquors.groupby('semana')['sale_usd','bottles_sold'].sum().reset_index()
#reset index es para q week no lo tome como índice y sea columna
#la diferencia de esta sintaxis que cuando ponemos .agg es que en agg como es un tipo diccionario, 
#  podemos decirle que en uno
#queremos suma y en el otro queremos media, por ejemplo 
```

```python
df_week = df_week[df_week['semana']>=45]
```

```python
df_week
```

```python
#Grafico de dispersión
plt.style.use('ggplot')
plt.figure(figsize=(20,10))
plt.title('Botellas por semana',fontsize =14)
plt.xlabel('Semana',fontsize=14)
plt.ylabel('Botellas vendidas',fontsize=14)
x = df_week['semana']
y = df_week['bottles_sold']
plt.scatter(x,y,s=y/1000,c=x,cmap='hot') #s es size y está determinado por y (las botellas) y x es color 
# que determina x (o sea la semana)... además lo hicimos/1000 porque si no salían enormes... y con cmap para determinar
#qué paleta quieres... puede ser vidris u otros. 
```

```python
#Qué botella es la que más tiene ventas... en lo siguiente sólo hacemos la tabla, para el top ponemos
#nlargest
df_botella = df_liquors.groupby('item_description')['sale_usd','bottles_sold'].sum().reset_index()
```

```python
df_botella.head()
```

```python
top_10_botellas = df_botella.nlargest(10, 'sale_usd') #con nlargest traemos el top 10 de botellas
```

```python
top_10_botellas
```

```python
#Grafico de linea

x = top_10_botellas['item_description']
y = top_10_botellas['sale_usd']
plt.plot(x,y, marker='.', linestyle ='dashed',markersize=12,color='blue')
plt.xticks(rotation='vertical')
plt.title('Top 10 botellas')
plt.xlabel('Botella')
plt.ylabel('Venta USD')

# plt.legend(x)
plt.savefig('test.jpg')
```

```python
# SUbplots
# Un lienzo para graficar en un mismo lugar
fig, axs = plt.subplots(2) #para dividir
axs[0].title.set_text('Ventas USD') #axs 0 es gráfico 1
axs[0].set_xticklabels([]) #quita primeras etiquetas porque corresponden a lo mismo las que están abajo
axs[0].plot(x,y)
y_botellas = top_10_botellas['bottles_sold']
plt.xticks(rotation='vertical')
axs[1].title.set_text('Botellas vendidas') #axs 1 es gráfico 2
axs[1].scatter(x,y_botellas)
```

```python
fig, axs = plt.subplots(1, 2, sharey=True)
axs[0].title.set_text('Ventas USD')
# axs[0].set_xticklabels([])
plt.xticks(rotation=90)
axs[0].plot(x,y)
y_botellas = top_10_botellas['bottles_sold']
plt.setp(axs[0].xaxis.get_majorticklabels(), rotation=90)
plt.xticks(rotation='vertical')
axs[1].title.set_text('Botellas vendidas')
axs[1].scatter(x,y_botellas)
```

```python
#Graficas de Pie
plt.style.use('classic')
colores = ["hotpink","red","green"]
explode = (.1,0,0,0,0,.1,0,0,.1,0)
plt.pie(y,labels = x, autopct ='%1.1f%%',colors=colores,startangle=90,explode=explode, shadow=True)
#autopct ='%1.1f%%' para que aparezca el porcentaje con un decimal
plt.show()
```

```python
#Barchart
plt.xticks(rotation='vertical')
plt.xlabel('Botellas top 10')
plt.ylabel('Ventas USD')
# plt.title('top 10 botellas')
plt.bar(x,y)
plt.show()
```

#Seaborn

```python
import seaborn as sns
```

```python
df_liquors.head(2)
```

```python
x_pack = df_liquors['pack']
y_precio_retail = df_liquors['state_bottle_retail']
```

```python
#Displot
sns.displot(x_pack,bins=10)
```

```python
sns.jointplot(x_pack,y_precio_retail,marginal_kws=dict(bins=10)) 
#con jointplot los unimos
#marginal hace bins para ver dónde hay mayor concentración

#Interpretación: tienes un pach como de dos.. y esas dos botellas te cuestan como 840 dólares
#mientras que la gráfica de arriba y la de al lado nos dice dónde se concentra mayor cantidad de packs
#aunque no veamos muchos puntos, vemos que la mayor concentración de paks está en de 10 a 15 botellas... 
#aunque no son los que más se venden... 
```

```python
df_corr = df_liquors[['bottle_volume','bottles_sold','sale_usd','pack']].corr()
cmap = sns.diverging_palette(6,255, as_cmap=True)
sns.heatmap(df_corr,cmap=cmap,annot=True)
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(x_pack,y_precio_retail, hue=x_pack,size=y_precio_retail) #hue nos ayuda a identificar
```

```python
sns.boxplot(x_pack,y_precio_retail)
```

```python
#Facet Grid
#revisar de nuevo desde min 1:54:00
df_pack = df_liquors.groupby('pack')['sale_usd'].sum().round(2).reset_index()
```

```python
df_pack.head(10)
```

```python
top_pack = df_pack.nlargest(5,'sale_usd')
top_pack['clase'] = 'top_5'
top_pack.head(10)
```

```python
df_liquors.head(2)
```

```python
table_concat = df_liquors.merge(top_pack[['pack','clase']],on=['pack'],how='left')
df_top = table_concat[table_concat['clase']=='top_5']
```

```python
multiple_lienzo = sns.FacetGrid(df_top,col='pack',row='date_mes')
multiple_lienzo.map(plt.hist,'sale_usd')
```
