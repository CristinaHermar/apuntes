```python
np.percentile(df_descriptive[['sale_usd','bottles_sold']],[25,75],axis=0)
```

```python
import numpy as np
q1, q3 = np.percentile(df_descriptive['bottles_sold'],[25,75],axis=0)

iqr = q3 - q1
upper = q3 + (iqr * 3)
lower = q1 - (iqr * 3)
```

```python
upper
```

```python
df_descriptive[df_descriptive['bottles_sold']>upper].index
```

```python
df_descriptive_clean = df_descriptive.drop(df_descriptive[df_descriptive['bottles_sold']>upper].index).reset_index()
```

```python
plt.figure(figsize=(20,10))
df_descriptive_clean.boxplot(column=['bottles_sold'])
```
