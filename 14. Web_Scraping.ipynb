```python
url = 'https://www.rottentomatoes.com/top/bestofrt/'

html = requests.get(url).content
```

```python
soup = BeautifulSoup(html, 'lxml')

# print(soup.prettify())
```

```python
links = soup.select('td a[class="unstyled articleLink"]') 
#voy a utilizar un tag de css para poder traerme la información. 
#pongo td por el table data y me traigo la clase, para esto le pido que 
#baje a y le defino qué ha con los corchetes.  
```

```python
links = soup.select('td a.unstyled.articleLink')
```

```python
links #no quiero que me lo traiga así por eso le pido que extraiga el texto en la siguiente
```

```python
title = [element.text for element in links]
```

```python
title
```

```python
title_clean = [ti.strip() for ti in title]
```

```python
title_clean
```

```python
reviews = soup.select('td[class="right hidden-xs"]')
```

```python
reviews_clean = [rev.text for rev in reviews]
```

```python
reviews_clean
```

```python
url = 'https://www.rottentomatoes.com'

links_to_movie = [url + links[i]['href'] for i in range(len(links))] 
#esto ya lo teníamos en 'links' arriba, le vamos a pedir el primer elemento, porque así está ordenado. 
```

```python
links_to_movie
```

```python
df = pd.DataFrame(list(zip(title_clean, reviews_clean, links_to_movie)),
                 columns=['Movie', 'Review', 'Links']) 
#zip junta las listas pero respetando el mismo nivel
```
